```python
# Letra A

import pandas as pd

df = pd.read_excel("pedidos_de_asilo.xlsx")

print(df.shape)
df.head()
```

```python
# Letra B

import matplotlib.pyplot as plt

df.hist(column='Número de Pedidos',bins = 100, grid=True)
plt.title("Distribuição do Número de Pedidos de Asilo")
plt.xlabel("Número de Pedidos")
plt.ylabel("Frequência")
plt.xlim(0,100)
plt.show()
```

```python
# Letra C
import matplotlib.pyplot as plt


plt.title("Número de Pedidos de Asilo ao Longo dos Anos")
plt.xlabel("Ano")
plt.ylabel("Número de Pedidos")
plt.grid(True)
plt.scatter(df['Ano'], df["Número de Pedidos"])
plt.show()
```
