```python
# Letra A

import pandas as pd

dados = {
    "Filial A": [7.8, 6.7, 2.8, 5.6, 8.3],
    "Filial B": [6.2, 4.1, 9.2, 4.5, 7.5],
    "Filial C": [9.8, 7.3, 3.4, 7.2, 9.1]
}

df = pd.DataFrame(dados, index=["Segurança", "Financeiro", "Transporte", "Tecnologia", "Marketing"])
df.index.name = "Área de Negócio"

df.head()
```

```python
# Letra B item 1

df["Media Desempenho"] = df.mean(axis=1)

df.head()
```

```python
# Letra B item 2

df['Melhor Desempenho'] = df.apply(lambda x: x.idxmax(), axis=1)

df.head()
```

```python
# Letra B item 3

soma = df.sum(axis=1) - df["Media Desempenho"]

print(soma)
```

```python
# Letra C item 1

df.to_csv("Desempenho.csv")
```
